```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]
```

```python
plt.plot(x, y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Basic line Plot")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y, color='blue', linestyle='--', marker='o', linewidth=3, markersize=9)

plt.grid(True)

plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [20, 35, 30, 25]

plt.bar(categories, values, color='skyblue', edgecolor='yellow')


plt.title("bar plot")
plt.xlabel("Categories")
plt.ylabel("Values")

plt.grid(True, axis='y')
plt.grid(True, axis='x')

plt.show()
```

```python
#create a scatter plot
x = [1,2,3,4,5,]
y = [10,20,30,40,50]
plt.scatter(x,y,color="green",marker='x')
```

```python
labels = ['A', 'B', 'C', 'D']
sizes = [30, 20, 40, 10]
colors = ['gold', 'yellowgreen', 'lightblue', 'lightcoral']
explode = (0.2, 0, 0, 0)  # explode first slice

plt.pie(sizes,labels=labels,colors=colors,explode=explode,autopct="%1.1f%%",shadow=True)


plt.show()
```

```python
labels = ['A', 'B', 'C', 'D']
sizes = [30, 20, 40, 10]
colors = ['gold', 'yellowgreen', 'lightblue', 'lightcoral']
explode = (0.2, 0, 0, 0)  # explode first slice

plt.pie(sizes,labels=labels,colors=colors,explode=explode,autopct="%1.1f%%",shadow=True,startangle=90)

plt.axis('equal')

plt.title("Pie Chart")


plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
#lineplot
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("line plot of total bill by size")
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
plt.title("bar plot of total bill by size")
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
sns.kdeplot(tips['total_bill'])
```
